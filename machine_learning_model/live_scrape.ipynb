```python
import socket
import struct
import numpy as np
import os

SAVE_FOLDER = "Dataset\class_1_hazard_unity"
MAX_FRAMES_TO_COLLECT = 100  

os.makedirs(SAVE_FOLDER, exist_ok=True)

UDP_IP = "127.0.0.1"
UDP_PORT = 5005
BUFFER_SIZE = 49152 

sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
sock.bind((UDP_IP, UDP_PORT))

print(f" Uncapped Harvester Armed. Listening to Unity...")

frame_count = 0

try:
    while frame_count < MAX_FRAMES_TO_COLLECT:
        
        data, addr = sock.recvfrom(BUFFER_SIZE)

        floats = struct.unpack(f'{len(data)//4}f', data)
        points = np.array(floats, dtype=np.float32).reshape(-1, 3)

        filename = os.path.join(SAVE_FOLDER, f"unity_hazard_few_{frame_count:04d}.npy")
        np.save(filename, points)
        
        frame_count += 1
        #
        print(f"\r Rapid Saving: {frame_count}/{MAX_FRAMES_TO_COLLECT} complete", end="")

except KeyboardInterrupt:
    pass

print("\nSafe Data collection complete!")
sock.close()
```
